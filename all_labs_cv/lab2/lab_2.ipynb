# Image Filtering in Python:

Image filtering is the process of applying a mathematical function to an image to modify it in some way. There are different types of filters available such as low-pass, high-pass, edge detection, etc. Here is an example code in Python for applying a 3x3 low-pass filter to an image using the OpenCV library:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('taylor.jpg')

def filter_image(img_path):
    # Load the image
    img = cv2.imread(img_path)

    # Define the kernel
    kernel = np.ones((3,3),np.float32)/9

    # Apply the filter
    dst = cv2.filter2D(img,-1,kernel)

    # Return the filtered image
    return dst
```

```python
filtered_img = filter_image('taylor.jpg')

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(filtered_img),plt.title('Output')
plt.show()
```

# Brightness and Contrast:
Brightness and Contrast are basic image processing operations that modify the intensity levels of an image. Here is an example code in Python using the OpenCV library:

```python
import cv2
import os

def bright_contr(folder_path, outPath):
    # Create the output folder if it doesn't exist
    if not os.path.exists(outPath):
        os.makedirs(outPath)

    # Create a loop to iterate through each image in the folder
    for filename in os.listdir(folder_path):
        # Load the image using cv2.imread() function
        img = cv2.imread(os.path.join(folder_path, filename))

        # Check if the image is loaded properly
        if img is not None:
            # Increase the brightness by 50
            bright_img = cv2.add(img, 50)

            # Decrease the contrast by 50
            contrast_img = cv2.subtract(img, 50)

            # Create new filenames for the adjusted images
            bright_name = os.path.splitext(filename)[0] + '_bright.jpg'
            contrast_name = os.path.splitext(filename)[0] + '_contrast.jpg'

            # Write the adjusted images to the output folder
            cv2.imwrite(os.path.join(outPath, bright_name), bright_img)
            cv2.imwrite(os.path.join(outPath, contrast_name), contrast_img)

            # Display the results
            plt.subplot(121),plt.imshow(img),plt.title('Input')
            plt.subplot(122),plt.imshow(bright_img),plt.title('bright_img')
            plt.show()
            print()
            plt.subplot(121),plt.imshow(img),plt.title('Input')
            plt.subplot(122),plt.imshow(contrast_img),plt.title('contrast_img')
            plt.show()
        else:
            print(f"{filename} could not be loaded.")
```

# Color Transformation:
Color Transformation is the process of converting an image from one color space to another. OpenCV supports several color spaces such as RGB, HSV, YCrCb, etc. Here is an example code in Python for converting an image from RGB to HSV using the OpenCV library:

```python
import cv2
import os

def rgb_to_hsv(inPath, outPath):
    # Create the output directory if it does not exist
    if not os.path.exists(outPath):
        os.makedirs(outPath)

    # Iterate through each image in the source folder
    for filename in os.listdir(inPath):
        # Load the image using cv2.imread() function
        img = cv2.imread(os.path.join(inPath, filename))

        # Check if the image is loaded properly
        if img is not None:
            # Convert the image from RGB to HSV
            hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

            # Create new filename for the transformed image
            new_filename = os.path.splitext(filename)[0] + '_hsv.jpg'

            # Save the transformed image to the output folder
            cv2.imwrite(os.path.join(outPath, new_filename), hsv_img)
            
            plt.subplot(121),plt.imshow(img),plt.title('Input')
            plt.subplot(122),plt.imshow(hsv_img),plt.title('color_transformation')
            plt.show()
            
        else:
            print(f"{filename} could not be loaded.")
```

# Histogram Equalization:
Histogram Equalization is a technique for adjusting the contrast of an image by modifying its intensity distribution. It can improve the overall contrast of an image and make details more visible. Here is an example code in Python for performing histogram equalization using the OpenCV library:

```python
import cv2
import os

def hist(inPath, outPath):
    
    if not os.path.exists(outPath):
        os.makedirs(outPath)

    # Iterate through each image in the source folder
    for filename in os.listdir(inPath):
        # Load the image using cv2.imread() function
        img = cv2.imread(os.path.join(inPath, filename))

        
        if img is not None:
            
            gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            
            # Apply histogram equalization
            eq_img = cv2.equalizeHist(gray_img)
            
            eq_img_file = os.path.splitext(filename)[0] + '_hist.jpg'

            # Save the transformed image to the output folder
            cv2.imwrite(os.path.join(outPath, eq_img_file), eq_img)
            
            plt.subplot(121),plt.imshow(img),plt.title('Input')
            plt.subplot(122),plt.imshow(eq_img),plt.title('Histogram Equalization')
            plt.show()
            
        else:
            print(f"{filename} could not be loaded.")
```

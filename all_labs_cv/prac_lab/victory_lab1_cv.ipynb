## Scaling

```python
def scaling():
    # Define the folder path containing the images
    folder_path = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/images'
    
    outPath = "C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/scaling"
    # Create a loop to iterate through each image in the folder
    for filename in os.listdir(folder_path):

        # Load the image using cv2.imread() function
        img = cv2.imread(os.path.join(folder_path, filename))

        # Apply the cv2.resize() function on the image
        res = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

        # Save the resized image with a new name in the desired folder
        new_filename = os.path.splitext(filename)[0] + 'scaled.jpg'
        cv2.imwrite(os.path.join(outPath, new_filename), res)

        # Display the resized image
        plt.subplot(121),plt.imshow(img),plt.title('Input')
        plt.subplot(122),plt.imshow(res),plt.title('Output')
        plt.show()
        
        #res.save(folder_path)
  
        #print(f"Saved image to {folder_path}")
    
scaling()
```

## Translation

```python
import os
import cv2
import numpy as np

def translation():
    # Define the folder path containing the images
    folder_path = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/images'
    
    # Define the path for the output folder
    outPath = "C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/Translation"

    # Create the output folder if it does not exist
    if not os.path.exists(outPath):
        os.makedirs(outPath)

    # Define the translation matrix
    M = np.float32([[1, 0, 100], [0, 1, 50]])

    # Create a loop to iterate through each image in the folder
    for filename in os.listdir(folder_path):

        # Load the image using cv2.imread() function
        img = cv2.imread(os.path.join(folder_path, filename), 0)

        # Apply the cv2.warpAffine() function on the image
        rows, cols = img.shape
        dst = cv2.warpAffine(img, M, (cols, rows))

        # Save the translated image with a new name in the desired folder
        new_filename = os.path.splitext(filename)[0] + '_translated.jpg'
        cv2.imwrite(os.path.join(outPath, new_filename), dst)

        # Display the translated image
        plt.subplot(121),plt.imshow(img),plt.title('Input')
        plt.subplot(122),plt.imshow(dst),plt.title('Output')
        plt.show()

translation()
```

## Rotation

```python
import os
import cv2 as cv

def rotation():
    # Define the source and destination folder paths
    src_folder = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/images'
    dst_folder = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/Rotation'

    # Check if the destination folder exists, and create it if not
    if not os.path.exists(dst_folder):
        os.makedirs(dst_folder)

    # Loop through each file in the source folder
    for filename in os.listdir(src_folder):
        # Load the image from the source folder
        img_path = os.path.join(src_folder, filename)
        img = cv.imread(img_path, 0)

        # Rotate the image by 90 degrees
        rows, cols = img.shape
        M = cv.getRotationMatrix2D(((cols-1)/2.0, (rows-1)/2.0), 90, 1)
        dst = cv.warpAffine(img, M, (cols, rows))

        # Save the rotated image to the destination folder with a new filename
        new_filename = os.path.splitext(filename)[0] + '_rotated.jpg'
        dst_path = os.path.join(dst_folder, new_filename)
        cv.imwrite(dst_path, dst)

        # Display the translated image
        plt.subplot(121),plt.imshow(img),plt.title('Input')
        plt.subplot(122),plt.imshow(dst),plt.title('Output')
        plt.show()

rotation()
```

## Affine Transformation

```python
import os
import numpy as np
import cv2 as cv

def aff_trans():

    folder_path = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/images'

    # Define the output folder path
    outPath = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/Aff_trans'

    # Create the output folder if it doesn't exist
    if not os.path.exists(outPath):
        os.makedirs(outPath)

    # Define the coordinates of the points in the original image
    pts1 = np.float32([[50,50],[200,50],[50,200]])

    # Define the coordinates of the points in the transformed image
    pts2 = np.float32([[10,100],[200,50],[100,250]])

    # Define the affine transformation matrix
    M = cv.getAffineTransform(pts1,pts2)

    # Create a loop to iterate through each image in the folder
    for filename in os.listdir(folder_path):

        # Load the image using cv2.imread() function
        img = cv.imread(os.path.join(folder_path, filename))

        # Apply the affine transformation using cv2.warpAffine() function
        rows,cols,ch = img.shape
        dst = cv.warpAffine(img, M, (cols,rows))

        # Save the transformed image with a new name in the desired folder
        new_filename = os.path.splitext(filename)[0] + 'affine.jpg'
        cv.imwrite(os.path.join(outPath, new_filename), dst)

        # Display the transformed image
        plt.subplot(121),plt.imshow(img),plt.title('Input')
        plt.subplot(122),plt.imshow(dst),plt.title('Output')
        plt.show()

aff_trans()
```

## Perspective Transformation

```python
import numpy as np
import cv2 as cv

def Perspective():

    folder_path = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/images'

    # Define the output folder path
    outPath = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/Perspec'

    # Create the output folder if it doesn't exist
    if not os.path.exists(outPath):
        os.makedirs(outPath)
        
    for filename in os.listdir(folder_path):

        # Load the image using cv2.imread() function
        img = cv.imread(os.path.join(folder_path, filename))

        # Apply the affine transformation using cv2.warpAffine() function
        rows,cols,ch = img.shape
        pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
        pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
        M = cv.getPerspectiveTransform(pts1,pts2)
        dst = cv.warpPerspective(img,M,(cols,rows))


        # Save the transformed image with a new name in the desired folder
        new_filename = os.path.splitext(filename)[0] + 'pers.jpg'
        cv.imwrite(os.path.join(outPath, new_filename), dst)

        # Display the transformed image
        plt.subplot(121),plt.imshow(img),plt.title('Input')
        plt.subplot(122),plt.imshow(dst),plt.title('Output')
        plt.show()
 
Perspective()
```

## Geometric Transformation 

```python
import cv2
import numpy as np

def Geometric():

    folder_path = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/images'

    # Define the output folder path
    outPath = 'C:/Users/Hamna Iftikhar/Downloads/COMPUTER_VISION/prac_lab/Geometric'

    # Create the output folder if it doesn't exist
    if not os.path.exists(outPath):
        os.makedirs(outPath)
        
    for filename in os.listdir(folder_path):
        
        img = cv.imread(os.path.join(folder_path, filename))

        # Apply the affine transformation using cv2.warpAffine() function
        num_rows, num_cols = img.shape[:2]
        
        rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), 30, 1)
        dst = cv2.warpAffine(img, rotation_matrix, (num_cols, num_rows))
        
        # Save the transformed image with a new name in the desired folder
        new_filename = os.path.splitext(filename)[0] + 'geometric.jpg'
        cv.imwrite(os.path.join(outPath, new_filename), dst)

        # Display the transformed image
        plt.subplot(121),plt.imshow(img),plt.title('Input')
        plt.subplot(122),plt.imshow(dst),plt.title('Output')
        plt.show()

Geometric()
```

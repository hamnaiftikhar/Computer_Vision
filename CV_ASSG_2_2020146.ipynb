```python
import cv2
import numpy as np

# read the input image
img = cv2.imread('taylor.jpg', cv2.IMREAD_GRAYSCALE)

# apply Gaussian smoothing to the image, Gaussian blurring is highly effective in removing Gaussian noise from an image.
blur = cv2.GaussianBlur(img, (3,3), 2)

# apply Canny edge detection to the blurred image
edges = cv2.Canny(blur, 100, 200)
```

```python
# show the resulting images
#original image, after converting it to grey scale
plt.imshow(img)
```

```python
# apply Gaussian smoothing to the image with kernal(3,3),it made the image more smoother by reducing the noise with deviation 2
plt.imshow(blur)
```

```python
#he Canny edge detection algorithm has identified the edges in the image and marked them in green type of lines.
plt.imshow( edges)
```

# TASK 2 (Pt. 2) USE OF SOBEL OPERATORS VS OTHER OPERATOR (Laplacian)

```python
import cv2
# Load the image
img = cv2.imread('taylor.jpg', 0)

prewittx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
prewitty = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
x = cv2.addWeighted(prewittx, 0.5, prewitty, 0.5, 0)

# Apply Scharr operator
scharrx = cv2.Scharr(img, cv2.CV_64F, 1, 0)
scharry = cv2.Scharr(img, cv2.CV_64F, 0, 1)
y = cv2.addWeighted(scharrx, 0.5, scharry, 0.5, 0)
# Apply Laplacian operator
v = cv2.Laplacian(img, cv2.CV_8U)
```

```python
plt.imshow(img)
```

# Result of sobel Operator

```python
plt.imshow(x)
```

# Result of Scharr operator

```python
plt.imshow(y)
```

##### Parameters for the Sobel() and Scharr() functions are similar to those used for the Sobel operator. The ksize parameter specifies the size of the kernel to be used for the gradient computation.

# Result of Laplacian operator

```python
plt.imshow(v)
```

###### The Laplacian operator detects edges based on the second derivative of the image intensity, which makes it more sensitive to noise and can produce more false positives than the Sobel and Scharr operators. whereas, sobel and scharr are less sensitive to noise than the Laplacian operator.

# Task 2 (Pt. 3)  Perfect edge detection vs Opencv canny algorithm

```python
import cv2

# Load the image
img = cv2.imread('taylor.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Experiment with different threshold values
for i in range(50, 200, 10):
    # Apply custom Canny edge detection
    edges = cv2.Canny(gray, i, i*3)

    # Display the Perfect Canny
    plt.imshow(edges)
```

```python
for i in range(50, 200, 10):
    # Apply OpenCV Canny edge detection
    edges_cv = cv2.Canny(gray, i, i*3, L2gradient=True)
    # Display the OpenCv Canny
    plt.imshow(edges_cv)
```

##### In this code,  first load the image and convert it to grayscale, then experiment with different threshold values for the Canny edge detector by varying the low and high thresholds from i to i*3 by using same edge detector and threshhold for better accuracy.
